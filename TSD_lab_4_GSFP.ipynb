$\sigma_1=0.3$, $\sigma_2=0.4$, $b_1=0.05$,
$b_2=0.03$, $r=0.07$, $S_1=130$, $S_2=100$, $T=0.25, 0.5$, $\rho=\{-0.5, 0, 0.5\}$ и $K=0.1, 0.2, \ldots, 1.0, 2.0, 3.0$, безрисковая ставка 7\%.

```python
sigma1 = 0.3
sigma2 = 0.4
b1 = 0.05
b2 = 0.03
r = 0.07
S1 = 130
S2 = 100
T_range = [0.25, 0.5]
rho_range = [-0.5, 0, 0.5]
K_range = list(np.arange(0.1, 3.1, 0.1))
```

```python
# оценка стоимости опциона колл на отношение цен двух активов
def quotient_option_price_Zhang(S1, S2, b1, b2, sigma1, sigma2, K, rho, T, is_call):
    theta = 1
    if is_call == False:
        theta = -1
    sigma_hat = np.sqrt(sigma1 ** 2 + sigma2 ** 2 - 2 * rho * sigma1 * sigma2)

    d1 = (np.log(S1 / S2 / K) + (b1 - b2 - 0.5 * (sigma1 ** 2 - sigma2 ** 2)) * T) / \
         (sigma_hat * np.sqrt(T))
    d2 = d1 + sigma_hat * np.sqrt(T)
    F = S1 / S2 * np.exp((b1 - b2 +  sigma2 * (sigma2 - rho * sigma1)) * T)
    option = theta * np.exp(-r * T) * (F * norm.cdf(theta * d2) - K * norm.cdf(theta * d1))

    return option
```

Для выполнения расчетов по приведенной формуле для двухактивного барьерного опциона необходимо:

1. **Определить необходимые параметры:**
   - \\( S_1, S_2 \\): текущие цены активов.
   - \\( H \\): барьер.
   - \\( X \\): цена исполнения опциона.
   - \\( T \\): срок до истечения опциона.
   - \\( \mu_1, \mu_2 \\): дрифтовые коэффициенты для активов.
   - \\( \sigma_1, \sigma_2 \\): волатильности активов.
   - \\( \rho \\): коэффициент корреляции между активами.
   - \\( \eta, \phi, b_1, r \\): дополнительные параметры, связанные с типом опциона.

2. **Расчет промежуточных параметров \\( d_1, d_2, d_3, d_4 \\):**
   Формулы для расчета представлены:
   $$
   d_1 = \frac{\ln(S_1 / X) + (\mu_1 + \sigma_1^2)T}{\sigma_1 \sqrt{T}}
   $$
   $$
   d_2 = d_1 - \sigma_1 \sqrt{T}
   $$
   $$
   d_3 = d_1 + \frac{2 \rho \ln(H / S_2)}{\sigma_2 \sqrt{T}}
   $$
   $$
   d_4 = d_2 + \frac{2 \rho \ln(H / S_2)}{\sigma_2 \sqrt{T}}
   $$

3. **Подставить значения в основную формулу и вычислить \( w \):**
   Основная формула:
   $$
   w = \eta S_1 e^{(b_1 - r)T} \left[ M(\eta d_1, \phi e_1; -\eta \phi \rho) - \exp \left( -\frac{2\mu_2 \ln(H / S_2)}{\sigma_2^2} \right) M(\eta d_3, \phi e_3; -\eta \phi \rho) \right]
   $$

   $$
   - \eta X e^{-rT} \left[ M(\eta d_2, \phi e_2; -\eta \phi \rho) - \exp \left( -\frac{2\mu_2 \ln(H / S_2)}{\sigma_2^2} \right) M(\eta d_4, \phi e_4; -\eta \phi \rho) \right]
   $$

4. **Создать графики зависимости стоимости опциона от переменных, например \\( S_1 \\), \\( S_2 \\) или \\( T \\).**

Добавлены формулы для расчета дополнительных параметров \\( e_1, e_2, e_3, e_4 \\) и уточнение для коэффициентов \\( \mu_1, \mu_2 \\):

### Формулы:
1. \\( e_1 \\):
   $$
   e_1 = \frac{\ln(H / S_2) - (\mu_2 + \rho \sigma_1 \sigma_2)T}{\sigma_2 \sqrt{T}}
   $$

2. \\( e_2 \\):
   $$
   e_2 = e_1 + \rho \sigma_1 \sqrt{T}
   $$

3. \\( e_3 \\):
   $$
   e_3 = e_1 - \frac{2 \ln(H / S_2)}{\sigma_2 \sqrt{T}}
   $$

4. \\( e_4 \\):
   $$
   e_4 = e_2 - \frac{2 \ln(H / S_2)}{\sigma_2 \sqrt{T}}
   $$

5. Коэффициенты \\( \mu_1 \\) и \\( \mu_2 \\):
   $$
   \mu_1 = b_1 - \frac{\sigma_1^2}{2}, \quad \mu_2 = b_2 - \frac{\sigma_2^2}{2}
   $$


```python
def call_down_and_out(S1, S2, T, sigma1, sigma2, r, b1, b2, X, H, rho):
    # Вычисление дрифтов
    mu1 = b1 - 0.5 * sigma1 ** 2
    mu2 = b2 - 0.5 * sigma2 ** 2

    # Вычисление d1, d2, d3, d4
    d1 = (np.log(S1 / X) + (mu1 + sigma1 ** 2) * T) / (sigma1 * np.sqrt(T))
    d2 = d1 - sigma1 * np.sqrt(T)
    d3 = d1 + 2 * rho * np.log(H / S2) / (sigma2 * np.sqrt(T))
    d4 = d2 + 2 * rho * np.log(H / S2) / (sigma2 * np.sqrt(T))

    # Вычисление e1, e2, e3, e4
    e1 = (np.log(H / S2) - (mu2 + rho * sigma1 * sigma2) * T) / (sigma2 * np.sqrt(T))
    e2 = e1 + rho * sigma1 * np.sqrt(T)
    e3 = e1 - 2 * np.log(H / S2) / (sigma2 * np.sqrt(T))
    e4 = e2 - 2 * np.log(H / S2) / (sigma2 * np.sqrt(T))

    # Функция для совместного нормального распределения
    def M(a, b, rho):
        return multivariate_normal.cdf([a, b], mean=[0, 0], cov=[[1, rho], [rho, 1]])

    # Вычисление стоимости опциона
    term1 = eta * S1 * np.exp((b1 - r) * T) * (
        M(eta * d1, phi * e1, -eta * phi * rho)
        - np.exp(2 * (mu2 + rho * sigma1 * sigma2) * np.log(H / S2) / sigma2 ** 2)
        * M(eta * d3, phi * e3, -eta * phi * rho)
    )

    term2 = eta * X * np.exp(-r * T) * (
        M(eta * d2, phi * e2, -eta * phi * rho)
        - np.exp(2 * mu2 * np.log(H / S2) / sigma2 ** 2)
        * M(eta * d4, phi * e4, -eta * phi * rho)
    )

    w = term1 - term2

    return w
```
