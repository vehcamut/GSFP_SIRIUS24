Постройте структурный продукт с полной гарантией возврата инвестированного капитала и азиатского опциона call.

```python
def Geometric_Average_Rate_Call(S, K , T , r , b , sigma):
  sigmaA = sigma/np.sqrt(3)
  bA = (b - sigma**2/6)/2
  d1 = (np.log(S/K)+(bA + sigmaA**2/2)*T)/(sigmaA*np.sqrt(T))
  d2 = d1 - sigmaA * np.sqrt(T)
  return S * np.exp((bA - r)*T) * norm.cdf(d1) - K * np.exp(-r*T)*norm.cdf(d2)

S = 100
K = 100
T = 1
r = 0.07
q = 0.0
sigma = 0.3

print ('Геометрическая среднее. Стоимость азиатского опциона колл =', Geometric_Average_Rate_Call(S , K , T , r , r-q , sigma))
# 7.906075677449117
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
S0 = 100  # Initial stock price
K = 100  # Strike price
r = 0.05  # Risk-free rate
sigma = 0.2  # Volatility
T = 1  # Time to maturity in years
n_simulations = 10000  # Number of simulations

# Time steps for different observation periods
steps_daily = int(252 * T)  # Daily (252 trading days per year)
steps_weekly = int(52 * T)  # Weekly (52 weeks per year)
steps_monthly = int(12 * T)  # Monthly (12 months per year)

# Simulate stock price paths using geometric Brownian motion
def simulate_paths(S0, r, sigma, T, steps, n_simulations):
    dt = T / steps
    paths = np.zeros((n_simulations, steps + 1))
    paths[:, 0] = S0
    for t in range(1, steps + 1):
        z = np.random.standard_normal(n_simulations)
        paths[:, t] = paths[:, t - 1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * z)
    return paths

# Calculate Asian option price and volatility for different observation periods
def calculate_asian_option_and_volatility(paths, K, r, T):
    average_price = np.mean(paths[:, 1:], axis=1)  # Exclude initial price
    payoff = np.maximum(average_price - K, 0)  # Call option
    option_price = np.exp(-r * T) * np.mean(payoff)
    vol = np.std(average_price)
    return option_price, vol

# Simulate paths and calculate metrics
paths_daily = simulate_paths(S0, r, sigma, T, steps_daily, n_simulations)
paths_weekly = simulate_paths(S0, r, sigma, T, steps_weekly, n_simulations)
paths_monthly = simulate_paths(S0, r, sigma, T, steps_monthly, n_simulations)

option_price_daily, vol_daily = calculate_asian_option_and_volatility(paths_daily, K, r, T)
option_price_weekly, vol_weekly = calculate_asian_option_and_volatility(paths_weekly, K, r, T)
option_price_monthly, vol_monthly = calculate_asian_option_and_volatility(paths_monthly, K, r, T)

# Collect results for comparison
results = {
    "Observation Period": ["Daily", "Weekly", "Monthly"],
    "Asian Option Price": [option_price_daily, option_price_weekly, option_price_monthly],
    "Volatility": [vol_daily, vol_weekly, vol_monthly]
}

# Plot results
x_labels = ["Daily", "Weekly", "Monthly"]
option_prices = [option_price_daily, option_price_weekly, option_price_monthly]
volatilities = [vol_daily, vol_weekly, vol_monthly]

# Plot Asian option price
plt.figure(figsize=(8, 5))
plt.bar(x_labels, option_prices, alpha=0.7)
plt.title("Asian Option Price by Observation Period")
plt.ylabel("Price")
plt.show()

# Plot volatility
plt.figure(figsize=(8, 5))
plt.bar(x_labels, volatilities, alpha=0.7, color='orange')
plt.title("Volatility by Observation Period")
plt.ylabel("Volatility")
plt.show()

import pandas as pd
results_df = pd.DataFrame(results)
import ace_tools as tools; tools.display_dataframe_to_user(name="Comparison of Asian Option Price and Volatility", dataframe=results_df)
```

```python
# Box-Muller transform to generate normal deviates
def getGaussian(mu, sigma, n):

	# Generate a pair of uniform random variables U,V
  U = np.random.uniform(0, 1, n)
  V = np.random.uniform(0, 1, n)

  # Get R~Exponential(1) and Theta~Uniform(0, 2pi) random variables
  R     = -np.log(U)
  Theta = 2*np.pi*V

	# Transform to X and Y
  X = np.multiply(np.sqrt(2*R), np.cos(Theta))
  Y = np.multiply(np.sqrt(2*R), np.sin(Theta))

  # Perform location-scale transform
  X = mu + sigma*X
  Y = mu + sigma*Y
  return X, Y
```

```python
def gbm_mcs(mu, sigma, S0, T, n, noOfPaths):
  S = []
  delta_t = T/n
  # Mean and variance of a lognormal random walk
  mean = S0 * np.exp((mu+(0.5*sigma**2))*T)
  variance = (S0 ** 2)*np.exp(2*mu*T + 2*(sigma ** 2)*T)*(np.exp(sigma**2)*T-1)
  for i in range(noOfPaths):
    X, Y = getGaussian(0, 1, n)
    dz_t = X
    dW_t = np.sqrt(delta_t) * dz_t
    dX_t = mu * delta_t + sigma * dW_t
    X_t = np.cumsum(dX_t)
    S_t = S0 * np.exp(X_t)
    S.append(S_t)
  S = np.array(S)
  return S, mean, variance
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Box-Muller transform to generate normal deviates
def getGaussian(mu, sigma, n):
    U = np.random.uniform(0, 1, n)
    V = np.random.uniform(0, 1, n)
    R = -np.log(U)
    Theta = 2 * np.pi * V
    X = np.sqrt(2 * R) * np.cos(Theta)
    X = mu + sigma * X
    return X

# Simulate Geometric Brownian Motion
def gbm_mcs(mu, sigma, S0, T, n, noOfPaths):
    delta_t = T / n
    S = np.zeros((noOfPaths, n))
    for i in range(noOfPaths):
        Z = getGaussian(0, 1, n)
        W_t = np.sqrt(delta_t) * Z
        X_t = mu * delta_t + sigma * W_t
        X_t = np.cumsum(X_t)
        S[i] = S0 * np.exp(X_t)
    return S

# Parameters
S0 = 100
noOfPaths = 1000
T = 1
n = 250
K = 100
r = 0.07
sigma = 0.3
mu = r - 0.5 * sigma ** 2

# Simulate the paths
S = gbm_mcs(mu, sigma, S0, T, n, noOfPaths)

# Monte Carlo calculation of the Asian option price
payoff = np.maximum(np.mean(S, axis=1) - K, 0)
option_price = np.exp(-r * T) * np.mean(payoff)
print(f'Монте-Карло моделирование. Стоимость азиатского опциона колл = {option_price:.2f}')

# Plot a sample of paths
time = np.arange(n)
step = 50
for i in range(0, noOfPaths, step):
    plt.plot(time, S[i])
plt.xlabel('Шаги по времени, $n$')
plt.ylabel('Базовый актив, $S$')
plt.grid(True)
plt.show()

# 3D Density Plot
step = 50
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(np.min(S), np.max(S), step + 1)
for t in range(step, n + 1, step):
    hist, bins = np.histogram(S[:, t - 1], bins=x, density=True)
    bins_center = (bins[:-1] + bins[1:]) / 2
    ax.bar(bins_center, hist, zs=t, zdir='y', alpha=0.8)

ax.set_xlabel('Базовый актив, $S$')
ax.set_ylabel('Время, $t$')
ax.set_zlabel('Частота')
plt.show()
```

```python
def bs_call_exact(t, s0, k, r, sigma):
    dp = 1 / ( sigma * math.sqrt( t ) ) * ( math.log( s0 / k ) + ( r + sigma**2 / 2) * ( t ) )
    dm = dp - sigma * math.sqrt( t )
    return s0 * norm.cdf(dp) - k * math.exp( - r * t ) * norm.cdf(dm)

# profit compare with vanilla call
def call(S, K, tau, vol, r, q):
    d1 = (np.log (S / K) + (r + vol**2 / 2) * tau) / (vol * np.sqrt(tau))
    d2 = d1 - vol * np.sqrt(tau)
    return S * np.exp(-q * tau) * norm.cdf(d1) - K * np.exp (-r * tau) * norm.cdf(d2)

@jit(nopython=True, inline = 'always')
def wiener(t):
    return math.sqrt( t ) * np.random.normal(0, 1)

@jit(nopython=True, inline = 'always')
def st(t, s0, r, sigma):
    return s0 * math.exp( ( r - sigma**2 / 2) * t + sigma * wiener( t ) )

@jit(parallel=True, nopython=False)
def asian_bs_call_mc_geom(n_iters, T, t_steps, S0, K, r, sigma):
    payoff = np.zeros(n_iters)
    avg_S = np.zeros(n_iters)
    m = len(t_steps)
    for i in prange(0, n_iters):
        geom_avg_s = 0
        S0_j = S0
        for j in range(1, m):
            S_j = st( (t_steps[j] - t_steps[j-1] ) * T, S0_j, r, sigma)
            S0_j = S_j
            geom_avg_s += math.log(S_j)
        geom_avg_s = math.exp(geom_avg_s / ( m - 1))
        avg_S[i] = geom_avg_s
        payoff[i] = math.exp(-r * T) * max(geom_avg_s - K , 0)
    return payoff, avg_S
```

```python
N_MC = int(10**5)
T, S0, K, r, sigma = 1, 100, 100, 0.07, 0.3
m = 1000
t_steps = np.array([i/m for i in range(0, m + 1)])
payoff_data, S_data = asian_bs_call_mc_geom(N_MC, T, t_steps, S0, K, r, sigma)
asian_call = np.mean(payoff_data)
print(f"asian call = {asian_call:.4f} +- {1.96 / math.sqrt(n) * np.std(payoff_data):.4f} with 95% prob")
```

# Задача

Сравните различные методы оценивания (арифметический и Монте-Карло) азиатского опциона для бездивидентных акций:
$S_0= 116,79$, $К= 117$, срок погашения составляет 3 месяца. Волатильность акций составляет 0,02803, а безрисковый процент составляет 0,624% с непрерывным начислением процентов.


https://github.com/olof98johansson/FinancialOptionsModelling

https://webthesis.biblio.polito.it/20278/1/tesi.pdf

```python
import numpy as np
from scipy.stats import norm

# Данные задачи
S0 = 116.79  # начальная цена акции
K = 117  # цена исполнения опциона
T = 3 / 12  # срок погашения (в годах)
sigma = 0.02803  # волатильность
r = 0.00624  # безрисковая ставка
n_simulations = 100000  # число симуляций для метода Монте-Карло
n_time_steps = 50  # число временных шагов для метода Монте-Карло

# Функция для аналитической оценки азиатского опциона с арифметическим средним
def analytical_asian_option(S0, K, r, T, sigma):
    """Расчет азиатского опциона с использованием аналитического метода."""
    sigma_adj = sigma * np.sqrt((2 * n_time_steps + 1) / (6 * (n_time_steps + 1)))
    r_adj = 0.5 * (r - 0.5 * sigma**2) + 0.5 * sigma_adj**2
    d1 = (np.log(S0 / K) + (r_adj + 0.5 * sigma_adj**2) * T) / (sigma_adj * np.sqrt(T))
    d2 = d1 - sigma_adj * np.sqrt(T)
    return np.exp(-r * T) * (S0 * norm.cdf(d1) - K * np.exp(-r_adj * T) * norm.cdf(d2))

# Метод Монте-Карло для оценки азиатского опциона
def monte_carlo_asian_option(S0, K, r, T, sigma, n_simulations, n_time_steps):
    """Расчет азиатского опциона с использованием метода Монте-Карло."""
    dt = T / n_time_steps
    discount_factor = np.exp(-r * T)
    simulated_payoffs = []

    for _ in range(n_simulations):
        prices = [S0]
        for _ in range(n_time_steps):
            z = np.random.normal()
            St = prices[-1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * z)
            prices.append(St)
        arithmetic_mean = np.mean(prices)
        payoff = max(0, arithmetic_mean - K)
        simulated_payoffs.append(payoff)

    return discount_factor * np.mean(simulated_payoffs)

# Расчеты
analytical_price = analytical_asian_option(S0, K, r, T, sigma)
monte_carlo_price = monte_carlo_asian_option(S0, K, r, T, sigma, n_simulations, n_time_steps)

analytical_price, monte_carlo_price
```

## Задача.

Имеется 0,1 млн рублей, которые клиент готов инвестировать на 1 год. Ожидается рост цен на акции «A».  Клиент готов нести 30\% потери в случае падении цены акции <<А>> (меньше чем при прямой покупке акции) в обмен на увеличение коэффициента участия. Текущая цена акции 100 руб.

Входные параметры продукта:
*   базовый актив -- акции,
*   срок -- $T = 1$ год,
*   цена начала участия -- $P_0 = 100$рублей,
*   барьерная цена (тип барьера) -- «включаемый» (knock-in) и «выключаемый» (knock-out),
*   условия наблюдения с целью определения пробития барьера -- частота
наблюдений, длина периода наблюдений, расположение периода наблюдений
в случае, если оно частичное – в начале, середине, конце срока продукта;
*   сценарий поведения базового актива -- рост.

Выходной параметр продукта:
*   коэффициент участия.


Проведите расчеты и вычислите коэффициент участия.

```python
import numpy as np
from scipy.stats import norm

# Входные параметры
S = 100  # текущая цена акции
K = 100  # страйк-цена
T = 1  # срок опциона в годах
sigma = 0.2  # волатильность (предполагаемая)
r = 0.05  # безрисковая ставка
barrier = 70  # барьер (knock-in, 30% ниже текущей цены)
investment = 100000  # сумма инвестиций

# Количество дискретных наблюдений
n_obs = 4  # квартальные наблюдения
dt = T / n_obs  # интервал между наблюдениями

# Расчёт стоимости европейского опциона call по формуле Блэка-Шоулза
def black_scholes_call(S, K, T, sigma, r):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# Вероятность пробития барьера (knock-in) с учётом дискретных наблюдений
def knock_in_probability(S, barrier, sigma, T, n_obs):
    p = 0
    for i in range(1, n_obs + 1):
        t = i * T / n_obs
        prob = norm.cdf((np.log(S / barrier) - (r - 0.5 * sigma**2) * t) / (sigma * np.sqrt(t)))
        p += prob
    return p / n_obs

# Стоимость барьерного опциона
def barrier_option_price(S, K, T, sigma, r, barrier, n_obs):
    vanilla_call = black_scholes_call(S, K, T, sigma, r)
    knock_in_prob = knock_in_probability(S, barrier, sigma, T, n_obs)
    return vanilla_call * knock_in_prob

# Расчёт стоимости барьерного опциона
barrier_option_cost = barrier_option_price(S, K, T, sigma, r, barrier, n_obs)

# Расчёт коэффициента участия (PR)
PR = K * barrier_option_cost / investment

print('Цена опциона: ', barrier_option_cost)
print('Коэффициент участия: ', PR * 100)
```
